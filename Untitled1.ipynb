```python
import requests
from datetime import datetime
from google.colab import drive
import csv

# 1. –ü–æ–¥–∫–ª—é—á–µ–Ω–∏–µ Google –î–∏—Å–∫–∞
drive.mount('/content/drive', force_remount=True)  # –ü–µ—Ä–µ–ø–æ–¥–∫–ª—é—á–µ–Ω–∏–µ

# 2. –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö —Å API –¶–ë –†–§
url = "https://www.cbr-xml-daily.ru/daily_json.js"
response = requests.get(url)
data = response.json()

# 3. –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –∫—É—Ä—Å–æ–≤ –≤–∞–ª—é—Ç
usd = data["Valute"]["USD"]["Value"]
eur = data["Valute"]["EUR"]["Value"]
cny = data["Valute"]["CNY"]["Value"]

# 4. –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print(f"üìÖ –î–∞—Ç–∞ –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è: {data['Date']}")
print(f"üíµ –î–æ–ª–ª–∞—Ä (USD): {usd} —Ä—É–±.")
print(f"üí∂ –ï–≤—Ä–æ (EUR): {eur} —Ä—É–±.")
print(f"üí¥ –Æ–∞–Ω—å (CNY): {cny} —Ä—É–±.")

# 5. –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤ CSV –Ω–∞ Google –î–∏—Å–∫
save_path = "/content/drive/MyDrive/currency_rates.csv"

with open(save_path, mode='w', encoding='utf-8', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["–í–∞–ª—é—Ç–∞", "–ö—É—Ä—Å", "–î–∞—Ç–∞"])
    writer.writerow(["USD", usd, datetime.now().strftime("%Y-%m-%d")])
    writer.writerow(["EUR", eur, datetime.now().strftime("%Y-%m-%d")])
    writer.writerow(["CNY", cny, datetime.now().strftime("%Y-%m-%d")])

print(f"\n‚úÖ –î–∞–Ω–Ω—ã–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤: {save_path}")
```
